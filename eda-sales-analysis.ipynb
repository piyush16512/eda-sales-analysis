```python
df.groupby('PRODUCTLINE')['SALES'].sum().sort_values().plot(kind='barh')
```

```python
df['SALES'].plot(kind='hist')
```

```python
sns.boxplot(x='DEALSIZE', y='SALES', data = df)
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%d/%m/%y')
monthly_sales = df.groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum()
monthly_sales.plot()
```

```python
plt.scatter(df['QUANTITYORDERED'], df['SALES'])
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
```
